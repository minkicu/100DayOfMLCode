```python
%matplotlib inline
import matplotlib.pyplot as plt
slices_hours = [4, 8]
activities = ['Sleep', 'Work']
colors = ['r', 'g']
plt.pie(slices_hours, labels=activities, colors=colors, startangle=90, autopct='%.1f%%')
plt.show()
```

```python
province_population = [1110, 987, 312, 97, 398]
activities = ['Dog', 'Cat', 'Rodent', 'Reptile', 'Fish']
plt.pie(province_population, labels=activities, startangle=90, autopct='%.2f%%')
plt.title('people in a certain town own certain types of pets')
plt.show()
```

```python
explodeTuple = (0.1, 0.0, 1.0, 0.0, 0.8)
colors  = ("red", "green", "orange", "cyan", "brown", "grey", "blue", "indigo", "beige", "yellow")

# Draw the pie chart
plt.pie(province_population, explode=explodeTuple,labels=activities,autopct='%1.2f',startangle=90,colors=colors)
plt.show()
```

```python
wedgeLabels     = ("Unigram", "Bigram", "Trigram", "Four-gram", "Five-gram", "Others")

ngramPercent    = (5, 5, 10, 5, 10, 65)
figureObject, axesObject = plt.subplots()

plt.pie(ngramPercent,
               labels       = wedgeLabels,
               shadow       = True,
               frame        = True,              
               startangle   = 120,
               autopct      = '%.1f%%',
               wedgeprops   = { 'linewidth' : 3,
                                'edgecolor' : "orange" })

plt.axis('equal')
plt.show()
```

```python
import numpy as np

np.random.seed(19680801)

N = 10
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)

ax = plt.subplot(111, projection='polar')
bars = ax.bar(theta, radii, width=width, bottom=0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Make data: I have 3 groups and 7 subgroups
group_names=['groupA', 'groupB', 'groupC']
group_size=[12,11,30]
subgroup_names=['A.1', 'A.2', 'A.3', 'B.1', 'B.2', 'C.1', 'C.2', 'C.3', 'C.4', 'C.5']
subgroup_size=[4,3,5,6,5,10,5,5,4,6]

# Create colors
a, b, c=[plt.cm.Blues, plt.cm.Reds, plt.cm.Greens]

# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors=[a(0.6), b(0.6), c(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')

# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names, labeldistance=0.7, colors=[a(0.5), a(0.4), a(0.3), b(0.5), b(0.4), c(0.6), c(0.5), c(0.4), c(0.3), c(0.2)])
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)

# show it
plt.show()
```
